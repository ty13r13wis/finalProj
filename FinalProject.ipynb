```python
#Import Libraries#
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load cleaned dataset
df = pd.read_csv('creditcardClean.csv')
```

```python
#Data cleaning to specify that count means counts of fraud#
df = df.rename(columns={'Class': 'Fraud'})
```

```python
#show clean data set
df
```

```python
# Calculate percentage of fraudulent transactions
fraud_percent = df['Fraud'].value_counts(normalize=True)[1] * 100
```

```python
# Create pie chart to display results
labels = ['Genuine', 'Fraudulent']
sizes = [100 - fraud_percent, fraud_percent]
explode = (0, 0.1)
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title('Percentage of Fraudulent Transactions')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the total amount for all cases
total_amount = df['Amount'].sum()

# Get the total amount for class 1 cases
class_1_amount = df[df['Fraud'] == 1]['Amount'].sum()

# Calculate the percentage of class 1 cases relative to the total amount
class_1_percent = class_1_amount / total_amount * 100

# Create a bar plot to show the percentage of class 1 cases relative to the total amount
plt.bar(['Fraud', 'Other'], [class_1_percent, 100 - class_1_percent])
plt.title('Percentage of Total Amount for Fraud Cases')
plt.ylabel('Percentage')
plt.show()
```
